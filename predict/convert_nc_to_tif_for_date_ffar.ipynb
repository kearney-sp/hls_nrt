```python
import rioxarray as riox
import xarray as xr
import os
import glob
import numpy as np
import datetime
```

```python
yr = 2023
inDIR = '../data/predictions/ffar_mi'
inWildcard = 'ffar_mi_biomass_kg_ha_' + str(yr) + '*.nc'
```

```python
outDIR = '../output/hls_biomass/ffar_mi'
if not os.path.exists(outDIR):
    os.mkdir(outDIR)
```

```python
fLIST = glob.glob(os.path.join(inDIR, inWildcard))
```

```python
xrLIST = []
for f in fLIST:
    xrLIST.append(riox.open_rasterio(f))#.sortby('y'))
```

```python
xr_all = xr.combine_by_coords(xrLIST)
```

```python
xr_all = xr_all.where(xr_all > 0, other=0)
```

```python
xr_all
```

```python
xr_all.sel(date=datetime.datetime(yr, 5, 1)).rio.to_raster(
    os.path.join(outDIR, 'ffar_mi_bm_kg_ha_' + str(yr) + '_5_1.tif'))
xr_all.sel(date=datetime.datetime(yr, 7, 15)).rio.to_raster(
    os.path.join(outDIR, 'ffar_mi_bm_kg_ha_' + str(yr) + '_7_15.tif'))
xr_all.sel(date=datetime.datetime(yr, 10, 31)).rio.to_raster(
    os.path.join(outDIR, 'ffar_mi_bm_kg_ha_' + str(yr) + '_10_31.tif'))
```
