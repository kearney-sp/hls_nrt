```python
import geopandas as gpd
```

```python
cper_f = 'data/ground/cper_pastures_2017_clip.shp'
cper = gpd.read_file(cper_f)
cper.loc[cper['Past_Name_'] == '21SW', 'Past_Name_'] = '21S'
cper_bbox = cper.buffer(500).total_bounds
#cper.geometry = cper.buffer(1)
cper = cper.dissolve('Past_Name_').reset_index()
idx_max = 1
while idx_max < 10:
    if any(cper.geometry.type == 'MultiPolygon'):
        print('trying')
        cper = cper.explode(index_parts=False)
        cper.geometry = cper.buffer(1*idx_max)
        cper = cper.dissolve('Past_Name_').reset_index()
        cper.geometry = cper.buffer(-1*idx_max)
        idx_max += 1
    else:
        break

cper['Pasture'] = cper.apply(lambda x: x['CARM_Name'] if x['CARM_Name'] is not None else x['Past_Name_'], axis=1)
```

```python
cper.to_file('data/ground/cper_pastures_2017_dissolved.shp', index=False)
```
